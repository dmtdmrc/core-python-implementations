```python
# Importing necessary libraries and packages 
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
import a1
from a1 import build_deep_nn
```

```python
# Determining number of hidden layers based on student id last digit:
student_id_ends = 0   # My student id: 48672130

if  student_id_ends in [0,3,6,9]:
    hidden_layers = 1
elif student_id_ends in [1,4,7]:
    hidden_layers = 2
else:
    hidden_layers = 3

# Determining size of the hidden layers based on student id last digit:
if student_id_ends in [0,1,2,3]:
    hidden_size = 100
elif student_id_ends in [4,5,6,7]:
    hidden_size = 75
else:
    hidden_size = 50

# Determining dropoout rate based on student id last digit:
if student_id_ends in [2,4,6]:
    dropout_rate = 0.3
elif student_id_ends in [0,3,7]:
    dropout_rate = 0.5
else:
    dropout_rate = 0.7
```

```python

input_size = 28 * 28
output_size =   10
layer_calc = [(hidden_size, 0)] *  (hidden_layers -1) + [(hidden_size, dropout_rate)]
    
```

```python
# Building build_deep_nn model 
model = build_deep_nn(input_size, layer_calc, output_size)
 
```

```python
# Defining function of loss and Adam_optimizer
loss_criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),lr = 0.001)

# Loading Mnist dataset
#Converting image to Pytorch_tensors 
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

# Mnist training and testing dataset
data_folder = "MNIST"
train_dataset = torchvision.datasets.MNIST(data_folder, download = True,  train = True, transform = transform )
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle = True)

test_dataset = torchvision.datasets.MNIST(data_folder, download = True,  train = False, transform = transform )
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
# Training the model

epochs = 5
for epoch in range(epochs):
    epoch_loss = 0.0
    for batch_images, batch_labels in train_loader:
        batch_images = batch_images.view(batch_images.shape[0], -1)
        optimizer.zero_grad()
        outputs = model(batch_images)
        loss_value=loss_criterion(outputs, batch_labels)
        loss_value.backward()
        optimizer.step()
        epoch_loss += loss_value.item()

    print(f" Epoch {epoch+1}, Loss: { epoch_loss / len(train_loader)}")
```

```python
# Determining the model

num_correct = 0
num_total= 0

with torch.no_grad():
    for batch_images, batch_labels in test_loader:
        batch_images = batch_images.view(batch_images.shape[0], -1)
        outputs = model(batch_images) 
        _, predicted = torch.max(outputs.data, 1)
        num_total += batch_labels.size(0)
        num_correct += ( predicted == batch_labels).sum().item()

Accuracy = 100* (num_correct / num_total)

print(f" Accuracy is {Accuracy:.2f}%")
```
